```python
#%pip install matplotlib
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
df = pd.read_csv(r'F:\Practice\Data Analytics\PYTHON for Data Analytics\19.Ice Cream Ratings.csv')
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use("dark_background")
```

```python
df.plot()
```

```python
df.plot(kind='line', title= "Ice Cream Ratings") 
#kind: Literal['line', 'bar', 'barh', 'hist', 'box', 'kde', 'density', 'area', 'pie', 'scatter', 'hexbin'
```

```python
df.plot(kind='line', title= "Ice Cream Ratings", xlabel='Daily Rating', ylabel='Score') 
```

```python
df['Flavor Rating'].plot(kind='bar', stacked=True) 
```

```python
df.plot(kind='barh', stacked=True) 
```

```python
df.plot(kind = 'scatter', x = 'Texture Rating', y = 'Overall Rating', c = 'red', s = 100)
```

```python
df.plot(kind = "hist", bins = 50)
```

```python
df.plot(kind="box")
```

```python
df.boxplot()
```
